# Simulation of generalized two cat code $F=(\hat{a}-\alpha(t))(\hat{a}-\beta(t))$

```python
%matplotlib inline
from qutip import *
from math import *
from pylab import *
import sys
import time
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
from matplotlib import cm
import cmath
import scipy.integrate
about()
```

# Input cat state of $\alpha_{in}=-\beta_{in}$

```python
# Space size
NN  = 35

# Time
ti  =  0 # Keep at zero always
tf  =  10
res =  10#10*tf
t=linspace(ti,tf,res) # start time, end time, number of increments

# State input
alps=-1.7
bets=1.7
inp=( coherent(NN,alps)+coherent(NN,bets) ).unit()

# NOTES: add Bloch vector for final state of qubit; and also for parallel movie of Wigner function
```

# Choose trajectory for $\alpha$ and $\beta$

```python
global alp,bet

# (1) Both alpha and beta shrinking and releasing; alps=2 bets=-2
#def alp(t): return alps-(0.25)*exp(-4*(t-2)**2)+0.*t
#def bet(t): return -alp(t)+0.*t

# (2) Vertical perturbation of alpha only: tf=4 res=80 alps=2 bets=-2
#def alp(t): return alps+1j*(0.25)*exp(-4*(t-2)**2)+0.*t 
#def bet(t): return bets+0.*t

# (3) Rotate beta around alpha
#def alp(t): return alps+0.*t
#def bet(t): return bets*(cos(t)-1j*sin(t))+0.*t

# (4) Closed path of alpha in both directions
#tf  =  600
res =  10*tf
t=linspace(ti,tf,res)
amp   = 0.35*tf
mu    = 0.5*tf
sigma = 1.*tf/6
def alp(t): return alps+0.*t 
def bet(t): return bets+0.*t + \
                        1.*amp/sqrt(2*pi*sigma**2)*exp(-0.5*(1.*(t-mu)/sigma)**2) + \
                        1j*amp*(t-mu)/sigma/sqrt(2*pi*sigma**2)*exp(-0.5*(1.*(t-mu)/sigma)**2)
#scipy.integrate.quad(lambda t: amp/sqrt(2*pi*sigma**2)*exp(-0.5*(1.*(t-mu)/sigma)**2), 0, tf)

# (5) Rotate both around origin
#def alp(t): return alps*(cos(t)+1j*sin(t))+0.*t
#def bet(t): return -alp(t)+0.*t

# (6) Rotate both around origin
#def alp(t): return alps+(0.5)*exp(-4*(t-2)**2)+1j*(2.)*(t-2)*exp(-4*(t-2)**2)+0.*t 
#def bet(t): return -alp(t)

# Sanity check: Vertical perturbation of beta only: tf=4 res=80 alps=2 bets=-2
#def alp(t): return alps+0.*t 
#def bet(t): return bets+1j*(0.25)*exp(-4*(t-2)**2)+0.*t

# Sanity check: Constants
#def alp(t): return alps+0.*t 
#def bet(t): return bets+0.*t

# Displacement of both and back
#def alp(t): return alps+1j*(0.5)*exp(-4*(t-2)**2)+0.*t
#def bet(t): return bets+1j*(0.5)*exp(-4*(t-2)**2)+0.*t

# Plot trajectories
frame = max(max(abs(bet(t))),max(abs(alp(t))))+0.05
plot(t,real(alp(t)),t,imag(alp(t)),t,real(bet(t)),t,imag(bet(t)))
legend(("Re(alpha)","Im(alpha)","Re(beta)","Im(beta)"))
ylim([-frame,frame]);
```

# Run Lindblad operator

```python
# Four Hamiltonian operators and their parameters as written in "arbitrary gate.lyx"
ham1 = -1j*(destroy(NN)-create(NN))
ham2 = destroy(NN)+create(NN)
ham3 = -1j*(destroy(NN)**2-create(NN)**2)
ham4 = destroy(NN)**2+create(NN)**2

# Four superoperators and their parameters as written in "arbitrary gate.lyx"
sup1 = 2*spre(destroy(NN))*spost(create(NN))-spre(num(NN))-spost(num(NN))
sup2 = 2*spre(destroy(NN)**2)*spost(create(NN)**2)-spre(num(NN)*(num(NN)-1))-spost(num(NN)*(num(NN)-1))
sup3 = 2*spre(destroy(NN))*spost(create(NN)**2)-spre(create(NN)*num(NN))-spost(create(NN)*num(NN))
sup4 = 2*spre(destroy(NN)**2)*spost(create(NN))-spre(num(NN)*destroy(NN))-spost(num(NN)*destroy(NN))

# Setup operator to measure coherence
pop = coherent(NN,alps)*coherent(NN,alps).dag()-coherent(NN,bets)*coherent(NN,bets).dag()
coh = coherent(NN,alps)*coherent(NN,bets).dag()

# Run
out=mesolve( [0.*qeye(NN),
              [ham1,real(alp(t)*bet(t)*conjugate(alp(t)+bet(t)))],
              [ham2,-imag(alp(t)*bet(t)*conjugate(alp(t)+bet(t)))],
              [ham3,-real(alp(t)*bet(t))],
              [ham4,imag(alp(t)*bet(t))]
              ],inp,t,
             [0.*qeye(NN),
              [sup1,absolute(alp(t)+bet(t))**2],
              [sup2,1.+0.*t],
              [sup3,-alp(t)-bet(t)],
              [sup4,-conjugate(alp(t)+bet(t))]
             ],[pop,coh],options=Options(store_states='True'));
```

# Show Wigner animation and initial state after that
